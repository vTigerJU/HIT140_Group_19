```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
df1 = pd.read_csv('dataset1.csv')
df2 = pd.read_csv('dataset2.csv')
df3 = pd.read_csv('dataset3.csv')

# Merge df1 (demographics) and df2 (screen time) on 'ID'
df = pd.merge(df1, df2, on='ID', how='inner')
```

```python
# Merge the result with df3 (well-being indicators) on 'ID'
dfm = pd.merge(df, df3, on='ID', how='inner')
```

```python
# Descriptive statistics for screen time on weekdays and weekends
screen_time_cols = ['C_we', 'C_wk', 'G_we', 'G_wk', 'S_we', 'S_wk', 'T_we', 'T_wk']
print(dfm[screen_time_cols].describe())
```

```python
# Descriptive statistics for well-being scores
wellbeing_cols = ['Optm', 'Usef', 'Relx', 'Intp', 'Engs', 'Dealpr', 'Thcklr', 'Goodme', 'Clsep', 'Conf', 'Mkmind', 'Loved', 'Intthg', 'Cheer']
print(dfm[wellbeing_cols].describe())
```

```python
# Ensure plots render inline in the notebook
%matplotlib inline

# Visualization 1: Screen Time (Weekend) vs Well-being Indicators
plt.figure(figsize=(10, 6))
plt.scatter(dfm['C_we'], dfm['Loved'], color='blue', label='Feeling Loved')
plt.scatter(dfm['C_we'], dfm['Conf'], color='green', label='Confidence')
plt.scatter(dfm['C_we'], dfm['Cheer'], color='red', label='Cheerfulness')
plt.xlabel('Screen Time on Weekends (C_we)')
plt.ylabel('Well-being Indicators')
plt.title('Screen Time (Weekend) vs Well-being Indicators')
plt.legend()
plt.show()
```

```python
# Visualization 2: Screen Time (Weekday) vs Well-being Indicators
plt.figure(figsize=(10, 6))
plt.scatter(dfm['C_wk'], dfm['Loved'], color='blue', label='Feeling Loved')
plt.scatter(dfm['C_wk'], dfm['Conf'], color='green', label='Confidence')
plt.scatter(dfm['C_wk'], dfm['Cheer'], color='red', label='Cheerfulness')
plt.xlabel('Screen Time on Weekdays (C_wk)')
plt.ylabel('Well-being Indicators')
plt.title('Screen Time (Weekday) vs Well-being Indicators')
plt.legend()
plt.show()
```

```python
# Visualization 3: Gender Distribution in the Dataset
plt.figure(figsize=(8, 5))
dfm['gender'].value_counts().plot(kind='bar', color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'], rotation=0)
plt.show()
```

```python
# Visualization 4: Minority Status vs Well-being Indicators
plt.figure(figsize=(10, 6))
plt.scatter(dfm['minority'], dfm['Loved'], color='blue', label='Feeling Loved')
plt.scatter(dfm['minority'], dfm['Conf'], color='green', label='Confidence')
plt.scatter(dfm['minority'], dfm['Cheer'], color='red', label='Cheerfulness')
plt.xlabel('Minority Status (0: Non-minority, 1: Minority)')
plt.ylabel('Well-being Indicators')
plt.title('Minority Status vs Well-being Indicators')
plt.legend()
plt.show()
```

```python
# Visualization 5: Screen Time Distribution (Weekend vs Weekday)
plt.figure(figsize=(10, 6))
plt.hist(dfm['C_we'], bins=20, alpha=0.5, label='Weekend Screen Time', color='blue')
plt.hist(dfm['C_wk'], bins=20, alpha=0.5, label='Weekday Screen Time', color='green')
plt.xlabel('Screen Time')
plt.ylabel('Frequency')
plt.title('Distribution of Screen Time (Weekend vs Weekday)')
plt.legend()
plt.show()
```
