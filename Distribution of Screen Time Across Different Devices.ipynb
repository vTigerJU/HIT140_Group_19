```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
dataset2 = pd.read_csv('dataset2.csv')
```

```python
df = dataset2.drop(['ID'], axis = 1)
df.describe()
```

```python
# Computer use
plt.figure(figsize=(14, 10))
plt.subplot(2, 2, 1)
plt.boxplot([dataset2['C_we'], dataset2['C_wk']], labels=['Weekends', 'Weekdays'])
plt.title('Screen Time on Computers')
plt.ylabel('Hours')
```

```python
# Video game use
plt.figure(figsize=(14, 10))
plt.subplot(2, 2, 2)
plt.boxplot([dataset2['G_we'], dataset2['G_wk']], labels=['Weekends', 'Weekdays'])
plt.title('Screen Time on Video Games')
plt.ylabel('Hours')
```
